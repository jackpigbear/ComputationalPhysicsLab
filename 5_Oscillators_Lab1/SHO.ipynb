**Euler Method**

```python
#令初位置＝震幅
k = 1
m = 1
omega = 1
theta = 80*((np.pi)/180) 
N = 2000
T=20 ##T是總共時間的演進，因此我們要設的較長，以確保能畫出完整的運動，多餘的部分後續再做刪除。
dt = T/N  #時間的間隔 也就是euler method 中的 delta t

def x_Time(T):
    x_0 = 1 
    x_1 = 1
    xf=[]
    for t in np.arange(0,T,dt):
        #x_1=dx/dt  etc..
        x_0 = x_0 + x_1*dt
        x_1 = x_1 - x_0*dt #位置問題互換？
        xf.append(x_0)
    return xf



xx = x_Time(T)

tt = np.linspace(0, T ,N)
xx = np.array(xx)
xx_ideal = np.cos(tt)

print(xx.size)
print(tt.size)
#落地時的y座標，理論上為趨近0

plt.plot(tt,xx)
plt.legend
plt.show()
#繪製運動軌跡



```
